## Visualizando Dados

A visualização de dados tem duas funções básicas: Explorar dados e Comunicar dados!

```python
from matplotlib import pyplot as plt
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
```

```python
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')
plt.title("PIB Nominal")
plt.ylabel("Billions of $")
plt.show()
```

### Gráficos de Barras

```python
movies = ["Fuga das galinhas", "Por água abaixo", "Pets", "Jojo Bizarre Adventures", "Nem que a vaga tussa"]
num_oscars = [5, 11, 3, 8, 10]
```

```python
plt.bar(range(len(movies)), num_oscars)
plt.title("My favorite moveis and animes")
plt.ylabel("# of Academy Awards")
plt.xticks(range(len(movies)), movies)     # rotular o eixo x com os nomes dos filmes
plt.show()
```

```python
from collections import Counter

grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 9]

histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)
```

```python
plt.bar([x+5 for x in histogram.keys()],
        histogram.values(),
        10,
        edgecolor=(0,0,0))

plt.xticks([10 * i for i in range(11)])     # rotula o eixo x em 0, 10, 20 ..., 100
plt.xlabel("Decile")
plt.ylabel("# of students")
plt.title("Distribuition of Exame 1 Grades")
plt.show()
```

```python
# o eixo y não iniciar em 0 pode desorientar os usuários (como mentir usando estatística)

mentions = [500, 505]
years = [2023, 2024]

plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("# of times i heard someone say something")
#plt.ticklabel_format(useOffset=False)
plt.axis([2022.5, 2024.5, 499, 506])
plt.title("Look at the 'Huge' Increase!")
plt.show()
```

```python
plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("# of times i heard someone say something")
plt.axis([2022.5, 2024.5, 0, 550])
plt.title("Not so Huge anymore!")
plt.show()
```

### Gráficos de Linhas

```python
# variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
# bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]

variance = [2 ** i for i in range(10)]
bias_squared = sorted(variance, reverse=True)

total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]
```

```python
plt.plot(xs, variance, 'g-', label='variance')
plt.plot(xs, bias_squared, 'r-', label='bias^2')
plt.plot(xs, total_error, 'b', label='total error')

plt.legend(loc=9)   # loc=9 significa 'top center'
plt.xlabel("model complexity")
plt.xticks([])
plt.title("The bias-variance tradeoff")
plt.show()
```

### Gráficos de Dispersão

```python
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
                 xy=(friend_count, minute_count),    # coloca o rotulo no respectivo ponto, mas levemente deslocado
                 xytext=(5,-5),
                 textcoords='offset points')

plt.title("Daily minutes vs. Number of friends")
plt.xlabel("# of friends")
plt.ylabel("daily minutes spent on the site")
plt.show()
```

```python
# se deixar para o matplotlib selecionar a escala talvez tenha uma comparação equivocada (mentir usando estatística)

grades_1 = [99, 90, 85, 97, 80]
grades_2 = [100, 85, 60, 90, 70]

plt.scatter(grades_1, grades_2)
plt.title("Axes arent comparable")
plt.xlabel("grades from test 1")
plt.ylabel("grades from test 2")
plt.show()
```

```python
# plt.axis("equal") traz uma dimensão melhor da situação com os eixos iguais

grades_1 = [99, 90, 85, 97, 80]
grades_2 = [100, 85, 60, 90, 70]

plt.scatter(grades_1, grades_2)
plt.axis("equal")
plt.title("Axes arent comparable")
plt.xlabel("grades from test 1")
plt.ylabel("grades from test 2")
plt.show()
```
